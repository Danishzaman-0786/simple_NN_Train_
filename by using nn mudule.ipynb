```python
import torch as t
```

```python
x_train_t,x_test_t,y_train_t,y_test_t=t.from_numpy(x_train),t.from_numpy(x_test),t.from_numpy(y_train),t.from_numpy(y_test)
```

```python
x_train_t
```

```python
import torch.nn as nn
```

```python
class Nn(nn.Module):
  def __init__(self,features):
    super().__init__()
    self.l1=nn.Linear(features,1)
    self.sig_=nn.Sigmoid()
  def forward(self,x):
    z=self.l1(x)
    y_pred=self.sig_(z)
    return y_pred
  def loss_(self,y,y_pred):
    epsilen=1e-7
    y_pred=t.clamp(y_pred,epsilen,1-epsilen)
    loss=-(y*t.log(y_pred)+(1-y)*(1-t.log(y_pred)))
    loss=loss.mean()
    return loss
```

```python
model=Nn(x_train_t.shape[1])
```

```python
x_train_t,x_test_t=x_train_t.to(t.float32),x_test_t.to(t.float32)
```

```python
model(x_train_t)
```

```python
summary(model,input_size=(x_train_t.shape))
```

```python
lr=0.0001
```

```python
epochs=50
```

```python
for i in range(epochs):
  #forward
  y_pred=model(x_train_t)
  #loss
  loss1=model.loss_(y_train_t,y_pred)
  #backpro..
  loss1.backward()
  with t.no_grad():
    model.l1.weight-=lr*model.l1.weight.grad
    model.l1.bias-=lr*model.l1.bias.grad
  # Zero gradients after update
  model.l1.weight.grad.zero_()
  model.l1.bias.grad.zero_()
  print(f'epoch:{i} loss:{loss1}')
```

```python
model(x_test_t)
```

```python
with t.no_grad():
  y_pred=model(x_test_t)
  y_pred=(y_pred>0.5).int()
print(y_pred)
```

```python
#accuracy
with t.no_grad():
  accuracy=(y_pred==y_train_t).float().mean()
print(f"accuracy:={accuracy}")
```
